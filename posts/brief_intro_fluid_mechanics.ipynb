I start with definitions from Wikipedia:

> Mechanics is the area of physics concerned with the relationship between force, matter, and motion among physical objects.

> Fluid mechanics is the branch of physics concerned with the mechanics of fluids (liquids, gases, and plasmas) and the forces on them.

Therefore, fluid mechanics basically studies how fluids behave when forces are applied.

In principle, we can describe a fluid’s behavior by tracking all of its particles over time. However, this is often impractical --- consider the [Avogadro constant](https://en.wikipedia.org/wiki/Avogadro_constant): about 18 gram of water contains approximately $6.02 \times 10^{23}$ molecules. For this reason, we usually treat a fluid as a continuous medium (also known as *continuum*) ignoring the discrete particles. Specifically in fluid mechanics, we assume that fluids consist of [fluid elements](https://en.wikipedia.org/wiki/Fluid_parcel) --- volumes that are small enough to be treated as points, yet large enough to contain many molecules. This allows us to define macroscopic properties such as density, pressure, and (bulk) velocity at each point in the fluid.

## Fluid Equations

Formally, we can *derive* the equations governing fluid behavior from the [Boltzmann equation](https://en.wikipedia.org/wiki/Boltzmann_equation) by taking *moments* of velocity. However, the basic fluid equations can be understood intuitively as conservation laws:

- **Mass conservation**: A volume that always contains the same particles cannot change its mass, although it may change its shape. This leads to the (mass) continuity equation:

$$
\frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \mathbf{u}) = 0
$$

Here, $\rho=\rho(\mathbf{r}, t)$ is the density and $\mathbf{u}=\mathbf{u}(\mathbf{r}, t)$ is the (bulk) velocity of the fluid element at position $\mathbf{r}$ and time $t$.

- **Momentum conservation**: A force represents the rate of change of momentum. In other words, momentum cannot change unless a force acts on the fluid element. This leads to the momentum equation:

$$
\rho\frac{d\mathbf{u}}{dt} = \mathbf{f}
$$

This is analogous to Newton's second law: 

$$
m\frac{d\mathbf{v}}{dt} = \mathbf{F}
$$ 

for a single particle with mass $m$ and velocity $\mathbf{v}$, where $\mathbf{F}$ is the force acting on the particle.

Since the density $\rho$ is mass per volume, $\mathbf{f}$ must be force per volume (also known as *body force*). Also, $\frac{d}{dt}$ in Newton's second law is derivative for **the** particle, then $\frac{d}{dt}$ in the momentum equation should be the derivative for **the** fluid element. This is known as the [material derivative](https://en.wikipedia.org/wiki/Material_derivative) $\frac{D}{Dt}$:

$$
\frac{d}{dt} = \frac{D}{Dt} = \frac{\partial}{\partial t} + \mathbf{u} \cdot \nabla
$$

Similar to the original Newton's second law, all the difficulties in the momentum equation arise from the body force $\mathbf{f}$. The most basic body force is the pressure gradient force $- \nabla p$, but gravity and viscous forces are also common. If we assume that the fluid is [Newtonian](https://en.wikipedia.org/wiki/Newtonian_fluid) with a Stokes hypothesis for the viscous force, we can write the momentum equation as:

$$
\rho\left(\frac{\partial \mathbf{u}}{\partial t} + \mathbf{u} \cdot \nabla \mathbf{u}\right) = - \nabla p + \rho \nu \nabla^2 \mathbf{u} + \frac{1}{3} \rho \nu \nabla (\nabla \cdot \mathbf{u}) + \mathbf{f}
$$

Here, $p=p(\mathbf{r}, t)$ is the pressure, $\nu$ is the kinematic viscosity (constant in this case), and $\mathbf{f}$ is the other body forces (e.g., gravity).


- **Energy conservation**: The (internal) energy of a system only change via heat transfer or work done on the system. This leads to the energy equation.

We can *derive* the energy equation from the first law of thermodynamics, $dU = \delta Q + \delta W$, but here we consider only adiabatic processes (i.e., no heat transfer, $\delta Q=0$). For such a case, the adiabatic condition $\frac{d}{dt}\frac{p}{\rho^\gamma}=0$ leads to:

$$
\frac{\partial p}{\partial t} = - \mathbf{u} \cdot \nabla p - \gamma p \nabla \cdot \mathbf{u}
$$

Here, $\gamma$ is the [adiabatic index](https://en.wikipedia.org/wiki/Heat_capacity_ratio).

## Summary of the Fluid Equations

### Continuity Equation (Mass Conservation)
$$
\frac{\partial \rho}{\partial t} = - \nabla \cdot (\rho \mathbf{u})
$$

### Momentum Equation (Momentum Conservation)
$$
\frac{\partial \mathbf{u}}{\partial t} = - \mathbf{u} \cdot \nabla \mathbf{u} - \frac{1}{\rho}\nabla p + \nu \nabla^2 \mathbf{u} + \frac{1}{3} \nu \nabla (\nabla \cdot \mathbf{u}) + \frac{1}{\rho}\mathbf{f}
$$

### Energy Equation (Energy Conservation)
$$
\frac{\partial p}{\partial t} = - \mathbf{u} \cdot \nabla p - \gamma p \nabla \cdot \mathbf{u}
$$

### Unknowns

The independent variables are position $\mathbf{r}$ and time $t$.

The quantities appearing in the equations are $\rho$, $\mathbf{u}$, $p$, $\mathbf{f}$, $\nu$, and $\gamma$.

Among them, $\mathbf{f}$, $\nu$, and $\gamma$ are typically given as constants or prescribed functions of position and time.

Therefore, the unknowns are $\rho$, $\mathbf{u}$, and $p$, which amount to five variables in total (1+3+1=5). Since the system also provides five equations (1+3+1=5), the fluid equations form a *closed* system.

## Navier–Stokes equations

The fluid equations above are also known as the [Navier–Stokes equations](https://en.wikipedia.org/wiki/Navier%E2%80%93Stokes_equations), which describe the behavior of [viscous fluid](https://en.wikipedia.org/wiki/Viscous_liquid) ($\nu \neq 0$). Often, the term "Navier–Stokes equations" refers to the momentum equation only, but here we consider the whole system of equations.

### Incompressible Fluid

If the density $\rho$ is constant **with respect to fluid elements**, i.e., $\frac{d\rho}{dt}=0$, the fluid is called [incompressible](https://en.wikipedia.org/wiki/Incompressible_flow).

From the continuity equation,

$$
\begin{align*}
\frac{\partial \rho}{\partial t} &= -\nabla \cdot (\rho \mathbf{u}) \\
&= -(\nabla\rho)\cdot\mathbf{u} - \rho\nabla\cdot \mathbf{u} \\
&= -\mathbf{u}\cdot\nabla\rho - \rho\nabla\cdot \mathbf{u} \\
\frac{\partial \rho}{\partial t} + \mathbf{u}\cdot\nabla\rho &= - \rho\nabla\cdot \mathbf{u} \\
\frac{d \rho}{d t} &= - \rho\nabla\cdot \mathbf{u} \\
\end{align*}
$$

Therefore, for incompressible fluids $\frac{d\rho}{dt}=0$, the continuity equation becomes $\nabla\cdot \mathbf{u} = 0$.

### Incompressible Navier–Stokes Equations

$$
\frac{\partial \mathbf{u}}{\partial t} = - \mathbf{u} \cdot \nabla \mathbf{u} - \frac{1}{\rho}\nabla p + \nu \nabla^2 \mathbf{u} + \frac{1}{\rho}\mathbf{f}
$$

$$
\nabla\cdot \mathbf{u} = 0
$$

Usually, we further assume that the density $\rho$ is uniform (note that incompressibility does not imply uniform density). In this case, the energy equation is not needed since the unknowns are now $\mathbf{u}$ and $p$ only, which amount to four variables in total (3+1=4). The system also provides four equations (3+1=4), so the incompressible Navier–Stokes equations form a *closed* system without the energy equation.
